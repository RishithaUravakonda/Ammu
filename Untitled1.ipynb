```
def largeDigit(n):
  i=0
  while n!=0:
    r=n%10
    if i<r:
      i=r
    n=n//10
  return i
n=int(input('enter the number'))
largeDigit(n)
```

```
def sumOfDigits(n):
  s=0
  while n>0:
    r=n%10
    s=s+r
    n=n//10
  return s
n=int(input('enter the number'))
sumOfDigits(n)
```

```
def palindrome(n):
  rev=0
  buffer=n
  while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
  if rev==buffer:
    print('palindrome')
  else:
    print('not a palindrome')
  return rev
n=int(input('enter the number'))
palindrome(n)
```

```
def large(n):
  i=0
  large=0
  while i!=3:
    r=n%10
    if large<r:
      large=r
    n=n//10
    i=i+1
  return large
n=int(input('enter the number'))
large(n)
```
