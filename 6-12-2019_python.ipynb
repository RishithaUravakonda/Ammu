```python
#include<stdio.h>
int main()
{
	char x;
	printf("enter the character");
	scanf("%c",&x);
	if(65<=x&&x<=90)
	{
		printf("uppercase character");
	}
	else if(97<=x&&x<=122)
	{
		printf("lowercase character");
	}
	else if(x==32)
	{
		printf("space character");
	}
	else if(48<=x&&x<=57)
	{
		printf("digital character");
	}
	return 0;
}
```

```python
#include<stdio.h>
int main()
{
	int x,sum,r;
	scanf("%d",&x);
	sum=0;
	while(x!=0)
	{
		r=x%10;
		if(r%2==0)
		{
			sum=sum+r;
		}
		x=x/10;
	}
	printf("%d",sum);
	return 0;
}
```

```python
#include<stdio.h>
int main()
{
	int small=9,large=0,x,r;
	scanf("%d",&x);
	while(x!=0)
	{
		r=x%10;
		if(small>=r)
		{
			small=r;
		}
		if(large<=r)
		{
			large=r;
		}
		x=x/10;
		
	}
	printf("smallest= %d""largest= %d",small,large);
	return 0;
}
```

```python
#include<stdio.h>
int main()
{
	int x,r,n,s=0;
	scanf("%d",&n);
	x=n;
	while(x!=0)
	{
		r=x%10;
		x=x/10;
		s=s*10+r;
	}
	if(s==n)
	{
		printf("palindrome");
	}
	else
	{
		printf("not a palindrome");
	}
	return 0;
}
```
