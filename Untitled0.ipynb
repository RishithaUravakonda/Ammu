datatype conversions


*   int()
         int() converts the input into int class
*   str()
          str() converts the input into string class

*   float()
          float() converts the input into float class






```
x="34.56"
print(x,end=' ')
print(type(x))
a=float(x)
print(a,end=' ')
print(type(a))
b=int(a)
print(b,end=' ')
print(type(b))
```

control flow statements

* conditional statements:
     -if else

*looping statements:
    - while
    - for





```
a=int(input('enter the number'))
if a%2==0:
  print('even number')
else:
  print('odd number')
```

```
a=int(input('enter your age'))
if 18<a<85:
  print('eligible to vote')
else:
  print('not eligible to vote')
```

```
a=10
b=3
print(a/b)#answer will be decimal value
print(a//b)#answer will be integer value
print(a%b)#modular division
print(a*10)#multiplication
print(a**10)#power
```
